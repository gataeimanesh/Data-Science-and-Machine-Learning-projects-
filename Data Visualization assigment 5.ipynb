## 4-Data Analysis


#### Q1-Which Race has hightest rate of heart diseases?

```python
plt.figure(figsize=(8, 6))
sns.countplot(x="Race", hue="HeartDisease", data=df, palette="pastel")
plt.title("Heart Disease by Race")
plt.xlabel("Race")
plt.ylabel("Count")
plt.legend(title="HeartDisease", labels=["No Heart Disease (0)", "Heart Disease (1)"])
plt.show()
```

```python
import pandas as pd

# Assuming you have a DataFrame called 'df' with columns "Race" and "HeartDisease"
# You can use pandas groupby and value_counts to count the occurrences of Heart Disease for each race

heart_disease_counts = df.groupby("Race")["HeartDisease"].value_counts().unstack().fillna(0)
heart_disease_counts["Total"] = heart_disease_counts.sum(axis=1)

# Find the race with the highest occurrence of Heart Disease
highest_race = heart_disease_counts["Total"].idxmax()

print("Race with the highest Heart Disease:", highest_race)
```

Q4- is there any relationship between two numerical variables BMI and SleepTime?

we present a  scatter plot which is  is used to visualise the relationship between two numerical variables, showing individual data points.

```python

plt.figure(figsize=(8, 6)) 
plt.scatter(df['BMI'], df['SleepTime'], s=50, c='blue', alpha=0.7, edgecolors='w') 
plt.xlabel('BMI')
plt.ylabel('SleepTime') 
plt.title('relationship between BMI and SleepTime') 
plt.grid(True) 
```

```python


# Ensure numeric data types for 'BMI' and 'SleepTime'
df['BMI'] = pd.to_numeric(df['BMI'], errors='coerce')
df['SleepTime'] = pd.to_numeric(df['SleepTime'], errors='coerce')

# Drop rows with missing values (if necessary)
df.dropna(subset=['BMI', 'SleepTime'], inplace=True)

# Create a correlation matrix
correlation_matrix = df[['BMI', 'SleepTime']].corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Set title and show the plot
plt.title('BMI vs. SleepTime Heatmap')
plt.show()
```
